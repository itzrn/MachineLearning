# Logistic Regression

Importing Module to read Dataset

```python
import pandas as pd
```

Importing Dataset

```python
dataset = pd.read_csv("Datasets/T10_Social_Network_Ads.csv")
X_features = dataset.iloc[:, :-1].values
y_outcome = dataset.iloc[:, [-1]].values
```

Printing X_features and y_outcome

```python
print("X_features\n", X_features)
print("y_outcome\n", y_outcome)
```

Printing the shape of X_features and y_outcome

```python
print("Shape of X_features ->", X_features.shape)
print("Shape of y_outcome ->", y_outcome.shape)
```

# Spliting the dataset into Training set and Test set

Importing train_test_split

```python
from sklearn.model_selection import train_test_split
```

Spliting the dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X_features, y_outcome, test_size=0.25, random_state=0)
```

Printing some data of X_train, X_test, y_train, y_test

```python
print("X_train -->\n", X_train[:10, :])
print("X_test --> \n", X_test[:10, :])
print("y_train --> \n", y_train[:10, :])
print("y_test -->\n", y_test[:10, :])
```

# Features Scalling

Importing StandardScaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

Applying StandardScaler to dataset

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Printing Scalered Data

```python
print("X_train -->\n", X_train[:10, :])
print("X_test --> \n", X_test[:10, :])
```

# Training The Logistic Regression Model on the Training Set

Importing Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

Creating Object of Logistic Rgression Class

```python
classifier = LogisticRegression(random_state=0)  #this is going to be a simple logistic Rgression
```

Training The Model On our Training Data

```python
classifier.fit(X_train, y_train.ravel())  # ravel() fun making y_train a 1D dataset
```

# Predicting a new result

Predicting For 30 and 87000

```python
 print(classifier.predict(sc.transform([[30, 87000]])))
```

Predicting For X_test

```python
print(classifier.predict(X_test))  # here no need to apply feature scalling, bcz its already scalled
```

Printing the predicted value and the test value in two column so we can compare the values


* importing numpy library

```python
import numpy as np
```

```python
y_pred = classifier.predict(X_test)
# np.set_printoptions(precision=2)  # this make the result with dedcimal place two
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test), 1)),1))
```

# Making Confusion Matrix

* Matrix showing how many mistakes and correct predictions are model made.
* we will compute the accuracy and we will see what is the percentage of correct predictions our model made on the test set.

Importing Confusion matrix module

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

Printing Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

65 correct predictions of the class zero meaning the customers of the test set who didn't buy the new SUV.
Then 24 correct predictions of the class one meaning correct predictions of the customers who bought the SUV. And then three incorrect predictions of the class one, meaning three incorrect predictions of the customers who, bought in reality the SUV bought were predicted not to. And finally, eight incorrect predictions of the class.
Zero meaning eight customers who in reality didn't buy the SUV but were predicted to buy it.

* Calculating Accuracy Classification Score

```python
acsc = accuracy_score(y_test, y_pred)
print(acsc)  # therefore our model is 89% accurate
```

And actually remember that there are 100 observations in a test set which means that we had indeed 89 correct predictions. Actually 65 here, plus 24 is equal indeed to 89. Alright! But for any size of the test set this would mean that you had 89% of correct predictions. And that's exactly the accuracy. The accuracy is the rate of correct predictions.

# Visualising The Training set result

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
